```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/data.csv')
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.hist(df['diagnosis'], color='blue')
plt.title('M=1 , B= 0')
plt.show
```

```python
#creating dummy var.
dummies = pd.get_dummies(df.diagnosis)
```

```python
#merging with original one
mrg = pd.concat([df, dummies], axis = 'columns')
#plt.figure(figsize=(20,20))
#sns.heatmap(mrg.corr(), annot=True)
#mrg
```

```python
#droping columns
mrg1 = mrg.drop(['diagnosis', 'Unnamed: 32', 'M'], axis = 'columns')
#mrg1
```

```python
mrg1.dtypes
```

```python
#conversion of data types
mrg1.iloc[:] = mrg1.iloc[:].astype('int')
```

```python
mrg1.dtypes
```

```python
#feature extraction
import numpy as np
feature_df=mrg1[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
x=np.asarray(feature_df)
y=np.asarray(mrg1['B'])
x
```

```python
#train/test spliting
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=6 , shuffle=True)
"""x_train.shape
y_train.shape
x_test.shape
y_test.shape"""
```

```python
#modeling
from sklearn import svm
classifier= svm.SVC(kernel='linear', gamma='auto', random_state=6)
classifier.fit(x_train, y_train)
y_predict=classifier.predict(x_test)
y_predict
```

```python
#Evaluation
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```
